# Preamble setup

## Introduction

The `hcn.preamble` module is a one‑line shortcut to a ready‑to‑use scientific Python environment.

## Usage

```python
from hcn.preamble import *
```

After this you have `np`, `plt`, `os`, `sleep`, `tqdm`, `e`, `pi` and the MATLAB‑style plotting preset available, without any further setup.

## Examples

````python
from hcn.preamble import *
````

````python
x = np.linspace(0, 2*pi, 128)
plt.plot(x, np.sin(x))
plt.show()
````

````python
x, y = np.meshgrid(np.linspace(-3, 3, 128), np.linspace(-3, 3, 128))
rho = np.sqrt(x**2 + y**2)

plt.imshow(np.exp(-rho**2))
plt.show()
````
