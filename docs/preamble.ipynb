# `hcn.preamble`

## Introduction

The `preamble` module is a one‑line shortcut to a ready‑to‑use scientific Python environment.

## Usage

```python
from hcn.preamble import *
```

After this you have `np`, `plt`, `os`, `sleep`, `tqdm`, `e`, `pi` and the MATLAB‑style plotting preset available, without any further setup.
